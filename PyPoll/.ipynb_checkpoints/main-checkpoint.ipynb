```python
import os
import csv
import statistics

#import original budget data csv file
election_csv = os.path.join("election_data.csv")

# Read in the CSV file
with open(election_csv, 'r') as csvfile:

    # Split the data on commas
    csvreader = csv.reader(csvfile, delimiter=',')
    headers = next(csvreader)
    #print(headers)
    
    #create empty tables for individual columns of voter_id, county and candidate data
    
    #initial vote counter = 0
    total_votes = 0
    
    #create empty table for list of candidates
    candidates = []
    
    #create table for each
    for row in csvreader:
        
        #count total number of votes as total number of rows in csv reader
        total_votes = total_votes + 1
        candidates.append(row[2])
    
    #use set function to find unique candidate strings in set of candidates
    candidates_unique = list(set(candidates))
        
    
        
    #create empty lists for total votes and percentages for each candidate
    candidate_votes = []
    candidate_percents = []
    
    #for loop to count total votes and calculate voter percentages
    for i in range(len(candidates_unique)):
        
        #create list for number of unique times a candidates name appears on the total candidates list = number of votes
        candidate_votes.append(candidates.count(candidates_unique[i]))
        
        #calculate percentage of each unique candidate votes
        candidate_perc_vote = candidate_votes[i]/total_votes
        
        #create list to store percentages associated with each unique candidate
        candidate_percents.append(round(candidate_perc_vote*100,3))
    
    #identify maximum number of votes as winning tally
    winner_votes = max(candidate_votes)
    
    print("Election Results")
    print("-"*70)
    print(f"Total Votes: {total_votes}")
    print("-"*70)
    
    #create zip table for candidates alongisde their votes and vote percentages and print sequence of name, percent and votes
    for name, votes, perc in zip(candidates_unique, candidate_votes, candidate_percents):
        print('{0}: {2}% {1}'.format(name, votes, perc))
    
    print("-"*70)
    
    for row in zip(candidates_unique, candidate_votes, candidate_percents):
        if row[1] == winner_votes:
            election_winter = row[0]
    
    print(f"Election Winner: {election_winner}")
    
    
        
```
